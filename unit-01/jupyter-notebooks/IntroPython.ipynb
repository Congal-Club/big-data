# Ejercicio de de Cálculo de la eficiencia general de equipos
###OEE=eficiencia_general(Overall Equipment Effectictiveness)
>### eficiencia_general= disponibilidad x eficiencia x calidad
>### disponibilidad= tiempo_disponivle- tiempo_muerto
>### eficienica= produccion_total/(tiempo_operativo x capacidad)
>### calidad=(produccion_total-defectos_y_reprocesos)/produccion_total




```python
produccion_total=45409        # Producción en el período registrado
capacidad=260                 # Capacidad de diseño por hora
piezas_defectuosas=3809       # Piezas con defecto en ciclo medido
tiempo_disponible=300-16      # tiempo total- tiempo planeado
tiempo_muerto=34+4            # tiempo por daños + tiempo por alistamientos y cambio de materiales
tiempo_operativo=284-38       # tiempo_disponible - tiempo_muerto
```

```python
disponibilidad=tiempo_operativo/tiempo_disponible
print(disponibilidad)
```

```python
eficiencia=produccion_total/(tiempo_operativo*capacidad)
print(eficiencia)
print(round(eficiencia,2))
```

```python
calidad=(produccion_total-piezas_defectuosas)/produccion_total
print(round(calidad,2))
```

```python
eficiencia_general=disponibilidad * eficiencia* calidad
print(round(eficiencia_general,2))
```

##Recursión

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n - 1)

print(factorial(0))
print(factorial(5))
print(factorial(10))

```

```python
#Reto 1
'''
Crear una función recursiva que permite imprimir los valores de 1 a n en orden ascendente.
Por ejemplo, si mi función tiene como nombre “imprimir” y al mandar llamar con un argumento numérico igual a 5,
 entonces imprimirá en pantalla los valores: 1, 2, 3, 4, 5.
'''
def imprimeAscendente(num):
  for valor in range(num):
    print(valor + 1)

imprimeAscendente(int(input("Valor que desea imprimir de manera ascendent")))
```

```python
#Acceder a elementos de tuplas
'''
Se puede acceder a los elementos de tuplas mediante la indexación y el slicing.
La indexación se usa para acceder a un solo elemento de una tupla o secuencia,
mientras que el slicing se usa para acceder a una subsecuencia de elementos.
La sintaxis para indexar y segmentar es la misma que para las listas. Aquí se muestran unos ejemplos:
'''
t = (1, 2, 3, 4, 5)
print(t[0])
print(t[-1] ) #Obtener el último valor de nuestra tupla
print(t[1:3])
```

```python
#Operaciones en tuplas
'''
Las tuplas se pueden usar de la misma manera que las listas para realizar operaciones como concatenación,
 repetición y pruebas de pertenencia. Sin embargo, debido a que las tuplas son inmutables, algunas operaciones
 como ordenar y agregar no están disponibles. Aquí se muestran unos ejemplos
'''
t1 = (1, 2, 3)
t2 = (4, 5, 6)
t3 = t1 + t2
print(t3) # (1, 2, 3, 4, 5, 6)
t4 = t1 * 5 #Repite t1 x veces
print(t4) # (1, 2, 3, 1, 2, 3, 1, 2, 3)
t5 = (1, 2, 3)
t6 = (0, 1, 2)
t7 = t5 - t6  # Esta operació - no es válida con n-tuplas
print(t7)
```

```python
#No se permite la modificación de los elemtnos, cada elemento se llama item
tupla = (1, 2, 3)
tupla[0] = 20
print(tupla)

```

```python
#Para modificar con rangos tampoco se puede utilizar con tuplas
tupla = (1, 2, 3, 4, 5, 6)
tupla[2:] = (20)
print(tupla)
```

```python
tupla = ("cadena1", "Cadena2", "Cadena3")
print(tupla)
```

```python
tupla = (23.23, False, "Cadena 1")
print(tupla)
```

```python
tupla = (23.23, 90, True, "Cadena 1")
print(tupla[3]) #Se indica la posición del elemento que deseamos nos muestre u obtener, tener cuidado cuando el índice no existe.
```

```python
#Métodos de longitud si se puede usar con tuplas
tupla = (1, 2, 3, 4, 5, 6)
print(len(tupla))
```

```python
#Union de tuplas
tupla1 = (1, 2, 3)
tupla2 = (2, 5, 6)
tupla3 = (7, 8, 9)
union_tuplas = (tupla1, tupla2, tupla3)
print(union_tuplas)
```

##Conjuntos

```python
numeros = {0, 1, 2, 3, 4, 5}
print(numeros)
#Agregar elemento al conjunto
numeros.add(6)
print(numeros)
numeros.update([7, 8, 9])
print(numeros)
```

```python
#Crear conjuntos de valores mixtos
conjunto_mixto = {20, "Esta es una cadena", (1, 2, 3)}
conjunto_mixto
```

```python
conjunto_enteros = {1, 2, 3, 3, 4, 4, 5}
conjunto_enteros
```

```python
#Generar conjunto con set--convierte una lista a conjunto
conjunto_lista = set([1, 2, 3, 4, 5, 5, 6, 6, 7])
conjunto_lista
```

```python
#Agregar elemento al conjunto con add y update
numeros = {1, 2, 3, 4, 5, 6}
numeros.add(6)
print(numeros)
numeros.update([7, 8, 9]) #Para agregar multiples valores
print(numeros)
```

```python
#Eliminar elementos de un conjunto con remove, discard
numeros = {1, 2, 3, 4, 5}
numeros.remove(4)  # El elemento a eliminar debe existir en el conjunto
print(numeros)
numeros.discard(6)# con discard elimina un elemento en específicos pero si no lo encuentra no pasa nada, no muestra error pero como remove si marcaría error.
print(numeros)
```

```python
estados={"Guanajuato", "Michoacán", "Jalisco", "Veracruz", "Sonora", "Puebla", "Durango", "Colima"}
print(estados)
for estado in estados:
  print(estado)
```

```python
#Operaciones con conjuntos
'''
Como se mencionó, los conjuntos son una colección de elementos únicos y no ordenados.
 Estos puedes ser utilizados para ejecutar operaciones de conjuntos tal como la unión, intersección y diferencia.
En el ejemplo, se crean dos conjuntos y se realizan las operaciones sobre estos:
'''
numeros1 = {1, 2, 3, 4}
numeros2 = {3, 4, 5}
print(numeros1.union(numeros2)) # {1, 2, 3, 4, 5}
print(numeros1.intersection(numeros2)) # {3}
print(numeros1.difference(numeros2)) # {1, 2}
```

```python
A = {"rojo", "verde", "azul"}
B = {"amarillo", "rojo", "naranja"}
print(A)
print(B)
#Unión de conjuntos
print(A.union(B))
print((A|B))

#Intersección entre conjuntos
print(A.intersection(B))
print((A&B))

#Diferencia entre conjuntos
print(A.difference(B))
print((A-B))
print((B-A))
```

```python
#Operaciones con los dicciondios como: agregar,modificar, eliminar,etc.
capitales = {"Jalisco": "Guadalajara", "Morelos": "Cuernavada"}
print(capitales)
#Agregar elementos al diccionario
capitales["Yucatán"] = "Merida"
print(capitales)

#Hacer modificaciones de elementos de una diccionario
capitales["Morelos"] = "Una prueba"
print(capitales)
capitales["Morelos"] = "Cuernavaca"
print(capitales)

#Eliminación de elementos en diccionarios
del capitales["Jalisco"]
print(capitales)

#Eliminar todos
capitales.clear()
print(capitales)
```

#Lectura y escritura sobre archivos csv

```python
datos=[
  ["Nombre", "Edad", "Ciudad"],
  ["Ignacio", "10", "Pénjamo"],
  ["Rosa", "51", "Celaya"],
  ["Jesus", "43", "Celaya"],
  ["María", "23", "Morelia"],
  ["José", "65", "Pénjamo"]
]

print(datos)
```

```python
#Escritura con writer sobre una archivo csv
import csv

with open("personas.csv", "w", newline = "") as file: #Wiht abre y cierra automaticamente el archivo cuando no se está usando.
  writer = csv.writer(file) #file es un apuntador
  writer.writerows(datos)

```

```python
#Lectura con csv.reader()
import csv

with open("personas.csv") as file:
  reader = csv.reader(file)
  for row in reader:
    print(row)
```

```python
#Lectura con DictReader() para leer los archivos csv como diccionarios
import csv
with open("personas.csv") as file:
  obj = csv.DictReader(file)
  for item in obj:
    print(item["Nombre"]+ " vive en "+ item["Ciudad"] + " y tiene "+ item["Edad"]+ " años.")
```

```python
'''
Tomando como base el ejercicio donde se generaron números aleatorios en un archivo,
crear un programa en Python que permita leer los números y realizar los siguientes cálculos:
media, mediana, moda y desviación estándar.
'''
import random

f = open('numeros2.text', 'w')

for count in range(100):
  num = random.randint(1, 1000)
  f.write(str(num)+' \n')
f.close()


```

```python
f = open('numeros2.text', 'r')
sumatoria = 0
print(count)
for linea in f:
  linea = linea.strip()
  numero = int(linea)
  sumatoria += numero
print(sumatoria)
print(sumatoria / count + 1)

#media
#mediana
#moda
#desviación estándar
f.close()


```
